```python
# Function to display the phonebook
def display_phonebook(phonebook):
    if phonebook:
        print("\nPhonebook:")
        for name, number in phonebook.items():
            print(f"{name}: {number}")
    else:
        print("Phonebook is empty.")

# Function to add a contact to the phonebook
def add_contact(phonebook):
    name = input("Enter the contact name: ")
    number = input("Enter the contact number: ")
    phonebook[name] = number
    print(f"Contact for {name} added successfully.")

# Function to search for a contact in the phonebook
def search_contact(phonebook):
    name = input("Enter the name to search: ")
    if name in phonebook:
        print(f"{name}: {phonebook[name]}")
    else:
        print(f"{name} not found in the phonebook.")

# Function to delete a contact from the phonebook
def delete_contact(phonebook):
    name = input("Enter the name to delete: ")
    if name in phonebook:
        del phonebook[name]
        print(f"Contact for {name} deleted successfully.")
    else:
        print(f"{name} not found in the phonebook.")

# Main program
def main():
    phonebook = {}  # Initialize an empty phonebook dictionary

    while True:
        print("\nPhonebook Menu:")
        print("1. Display Phonebook")
        print("2. Add Contact")
        print("3. Search Contact")
        print("4. Delete Contact")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            display_phonebook(phonebook)
        elif choice == '2':
            add_contact(phonebook)
        elif choice == '3':
            search_contact(phonebook)
        elif choice == '4':
            delete_contact(phonebook)
        elif choice == '5':
            print("Exiting the phonebook.")
            break
        else:
            print("Invalid choice, please try again.")

# Run the main program
if __name__ == "__main__":
    main()
```
